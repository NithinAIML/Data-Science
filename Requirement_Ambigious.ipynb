```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.cluster.util import cosine_distance
import numpy as np
import pandas as pd
import networkx as nx
import operator
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')


def read_article(file_name):
    df = pd.read_excel(file_name)
    features = df['Description'].tolist()
    #     print(features)
    sentences = [feature.split(".") for feature in features]
    print(sentences)

    return sentences, df


def sentence_similarity(sent1, sent2, stopwords=None):
    if stopwords is None:
        stopwords = []

    sent1 = [w.lower() for w in sent1]
    sent2 = [w.lower() for w in sent2]

    all_words = list(set(sent1 + sent2))

    vector1 = [0] * len(all_words)
    vector2 = [0] * len(all_words)

    # build the vector for the first sentence
    for w in sent1:
        if w in stopwords:
            continue
        vector1[all_words.index(w)] += 1

    # build the vector for the second sentence
    for w in sent2:
        if w in stopwords:
            continue
        vector2[all_words.index(w)] += 1

    return 1 - cosine_distance(vector1, vector2)


def build_similarity_matrix(sentences, stop_words):
    # Create an empty similarity matrix
    similarity_matrix = np.zeros((len(sentences), len(sentences)))

    for idx1 in range(len(sentences)):
        for idx2 in range(len(sentences)):
            if idx1 == idx2:  # ignore if both are same sentences
                continue
            similarity_matrix[idx1][idx2] = sentence_similarity(sentences[idx1], sentences[idx2], stop_words)

    return similarity_matrix


def status_by_pos(zip_of_tag_frequency):
    noun = ['Noun', 'NN', 'NNS', 'NNP', 'NNPS']
    pronoun = ['Pronoun', 'PRP', 'PRP$', 'WP$']
    adverb = ['Adverb', 'RB', 'RBR', 'RBS', 'RP', 'WRB']
    verb = ['Verb', 'VB', 'VBD', 'VBG', 'VBN', 'VBP', 'VBZ']
    adjective = ['Adjective', 'JJ', 'JJR', 'JSS']

    main_pos = []
    main_pos.append(noun)
    main_pos.append(pronoun)
    main_pos.append(adverb)
    main_pos.append(adjective)
    main_pos.append(verb)

    dic = {'Noun': 0, 'Pronoun': 0, 'Verb': 0, 'Adverb': 0, 'Adjective': 0}
    for i, j in zip_of_tag_frequency:
        for pos in main_pos:
            if i in pos:
                dic[pos[0]] += j
                break

    list = [(k, v) for k, v in dic.items()]
    max_pos = max(dic.items(), key=operator.itemgetter(1))[0]
    status = {'Verb': 'Then', 'Adjective': 'Given'}
    if max_pos in status:
        output = status[max_pos]
    else:
        output = "When"
    return list, output


def append_to_excel(summerized_list, pos_tag, pos_count_list, common_highest_pos, status, df):
    se = pd.Series(summerized_list)
    pos_series = pd.Series(pos_tag)
    pos_count_series = pd.Series(pos_count_list)
    common_highest_pos_series = pd.Series(common_highest_pos)
    status_series = pd.Series(status)
    df['Summerized_text'] = se.values
    df['Pos-Tag'] = pos_series.values
    df['Pos-Tag-Frequency'] = pos_count_series.values
    df['Highest_Pos_Tag'] = common_highest_pos_series.values
    df['Status'] = status_series.values

    print(df)
    df = df.loc[:, ~df.columns.str.contains('^Unnamed')]
    df.to_excel('summary.xlsx')
    return "Sucessfully Updated"


def generate_summary(file_name):
    nltk.download("stopwords")
    stop_words = stopwords.words('english')

    # Step 1 - Read text anc split it
    sentences, df = read_article(file_name)
    num = 1
    summerized_text_list = []
    pos_tag = []
    pos_count_list = []
    common_highest_pos = []
    status = []
    for sentence in sentences:
        # Step 2 - Generate Similary Martix across sentences
        print(num)
        num = num + 1
        summarize_text = ''

        sentence_similarity_martix = build_similarity_matrix(sentence, stop_words)
        sentence_similarity_martix = np.nan_to_num(sentence_similarity_martix)
        # Step 3 - Rank sentences in similarity martix
        sentence_similarity_graph = nx.from_numpy_array(sentence_similarity_martix)

        scores = nx.pagerank(sentence_similarity_graph, max_iter=1000)
        print('scores', scores)

        # Step 4 - Sort the rank and pick top sentences
        ranked_sentence = sorted(((scores[i], s) for i, s in enumerate(sentence)), reverse=True)
        print("Indexes of top ranked_sentence order are ", ranked_sentence)

        if len(ranked_sentence) > 0:
            for i in range(len(ranked_sentence)):
                summarize_text = summarize_text + " " + ranked_sentence[i][1]
        else:
            summarize_text = summarize_text + "No Summerized Text Found"

        text = word_tokenize(summarize_text)
        pos = nltk.pos_tag(text)
        # print(pos)
        pos_count = [j for i, j in pos]
        # print(pos_count)
        pos_count_series = pd.Series(pos_count)
        series_index = pos_count_series.value_counts().index.tolist()
        # print(series_index)
        series_value = pos_count_series.value_counts().tolist()
        # print(series_value)
        zip_count = list(zip(series_index, series_value))

        print(zip_count)
        highest_pos, output = status_by_pos(zip_count)

        common_highest_pos.append(highest_pos)
        status.append(output)
        pos_count_list.append(zip_count)
        pos_tag.append(pos)
        summerized_text_list.append(summarize_text)
        # Step 5 - Offcourse, output the summarize texr
        print("Summarize Text: \n", summarize_text)
    status = append_to_excel(summerized_text_list, pos_tag, pos_count_list, common_highest_pos, status, df)
    print(status)


# let's begin
generate_summary("summary.xlsx")
```
